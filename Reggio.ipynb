```python
ClasseEmissReggio = pd.read_csv("ClasseEmissReggio.csv", sep=';' )
```

```python
Result = pd.merge(ReggioModello, ReggioMarche, how='left', on='MarcaID')
Result
```

```python
Result2 = pd.merge(AutoReggio, Result, how='left',  on=['MarcaID', 'ModelloID'])
Result2
```

```python
del Result2['Societa_x']
del Result2['Societa_y']
del Result2['ValoreInventario']
```

```python
Result3 = pd.merge(Result2, ClasseEmissReggio, how='left',  on=['AutomezzoID'])
Result3
```

```python
Result3.drop([199,200], inplace = True)
Result3.reset_index(drop=True, inplace = True)
Result3
```

```python
del Result3['Societa_y']
del Result3['V_9_ClasseEmissione']
del Result3['PrimaRevisione']
del Result3['RevisioniSuccessive']
del Result3['ClassificazioneAutoRegioneId']

Result3.rename(columns={"Descrizione_y": "Marca", "Descrizione_x": "Modello", "Societa_x": "Societa"}, inplace=True)
Result3
```

```python
columnsTitles = ['Societa','AutomezzoID','MarcaID','Marca','ModelloID','Modello','Posti','TipoVeicoloID','TipoVeicolo','TipoCarburanteID','TipoCarburante','ConsumoMedio','ClasseEmissione','EuroEmissione','EuroIDEmissione','DataDiAcquisto','Immatricolazione','Chilometraggio','ClassePatente','Rottamato','DataRottamazione','KmTagliando']

Result3 = Result3.reindex(columns=columnsTitles)
```

```python
Result3.isna().sum()
```

```python
Result3['Immatricolazione'].fillna(method='ffill', inplace = True)
Result3
```

```python
Result3['Immatricolazione'] = Result3['Immatricolazione'].astype('datetime64')
Result3['Immatricolazione']
```

```python
#pd.set_option('display.max_rows', Imm.shape[0]+1)
```

```python
df = pd.DataFrame(Imm.Data_time.dt.strftime('%Y-%m-%d').str.split('-').tolist(),columns=['AnnoImmat','Month','Day'],dtype=int)
df
```

```python
del df['Month']
del df['Day']
```

```python
CarReggio = pd.concat([Result3, df], axis=1, sort=False)
del CarReggio['Immatricolazione']
CarReggio
```
